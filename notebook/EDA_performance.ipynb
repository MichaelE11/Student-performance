# About the Data
- This data investigates the achevements of students in two schools Gabrel perera (GP) or Malsnho da Seliviera. Durng data gatherng two dataset was collected for two subjects. This includes mathematiics and portuguise language

# Varables
- School GP or MS
- Sex    M or F
- Age 15 to 22
- Address urban(U) or Rural (R)
- Family size(famsize) Less or equal to 3 (LE3) or Greater or equal to 3(GE3).
- Pstatus(parental cohabitation status) together(T) or Apart(A)
- Medu (Mothers Educaton) none-0, prmary-1, 4th grade-2, 5th to 9th grade-3, secondary education-4
- Fedu same as Medu
- Mjob (Mothers Job) Teacher, healthcare services, civil servant, at home or others
- Fjob (Fathers job) same as Mjob
- reason (reason for attending the school) close to home, school reutation, course performance or others
- Guidian mother(M) or father(F)
- Traveltime (the tme student travel from home to school) 1-< 15min, 2- 15 to 30min, 3- 30min to 1hr, 4->1hr
- study time 1-<2hrs, 2-2 to 5 hrs, 3-5 to 10hrs, 4->10hrs
- failures (number of past failures) n: if 1<=n<3 else 4
- Schoolsup (extral educational support) yes or no
- famup (extral family educational support) yes or no
- paid (extral paid classes within the course subject (math or portugise) yes or no
- Activity(extral curricular activity) yes of no
- Nursery (attended nursery school) yes or no)
- higher(wants to take higher education) yes or no
- internet (have internet access at home?) yes or no
- Romantic (with a romantic relationship) yes or no
- famrel (quality of family relationship) 1 to 5 with 5 as the highest quality
- free time (free time after school ) from 1 to 5 with 5 as the highest time after school
- Goout (go out with friends) from 1 to 5
- Dalc (workday alcohol consumption) from 1 to 5
- Walc (weekend alcohol consumption) from 1 to 5
- health (current health status) from 1 to 5
- Absence from 0 to 93
- G1 (first period grade) frmo 0 to 20
- G2 (second period grade) from 0 to 20
- G3 (final period grade) from 0 to 20
